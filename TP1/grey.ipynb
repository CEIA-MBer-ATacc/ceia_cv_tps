```python
import os
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def show_hist(img):


    # Graficar
    plt.figure(figsize=(12, 5))

    # Calcular histogramas con np.histogram
    hist, bins = np.histogram(img.ravel(), bins=256, range=(0, 256))
    ax1=plt.subplot(221)
    ax1.plot(hist, label='256 bins')
    
    # Calcular histogramas con np.histogram
    hist, bins = np.histogram(img.ravel(), bins=128, range=(0, 256))
    ax2=plt.subplot(222)
    ax2.plot(hist, label='128 bins')

    # Calcular histogramas con np.histogram
    hist, bins = np.histogram(img.ravel(), bins=64, range=(0, 256))
    ax3=plt.subplot(223)
    ax3.plot(hist, label='64 bins')

    plt.title("Histogramas en gris")
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Ajuste para que no se superponga el t√≠tulo
    plt.grid(True)
    plt.show()
```

```python
# Extensiones permitidas
extensiones = ('.png')

# Iterar sobre los archivos del directorio actual
for archivo in os.listdir('.'):
    if archivo.lower().endswith(extensiones):
        print(f"Procesando: {archivo}")
        img = cv.imread(archivo, cv.IMREAD_GRAYSCALE)

        if img is None:
            print(f"‚ö†Ô∏è No se pudo leer {archivo}")
            continue
        
        # Mustro la imagen original
        plt.figure()
        plt.imshow(img, cmap='gray', vmin=0, vmax=255)
        plt.show()

        #Mostramos el histograma de la imagen original
        show_hist(img)
```
